```python
import numpy as np
import pandas as pd
df = pd.read_csv('customer.csv')
df.sample(5)
```

```python
df = df.iloc[:,2:]
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop('purchased',axis=1),df.purchased,random_state=0,test_size=0.3)
x_train.shape 
y_test
```

```python
# ordinal encoder encodes based on the weight or importance of categories. ex-good is more and poor is less important,weight
from sklearn.preprocessing import OrdinalEncoder
x_train
```

```python
# this encoding is given in order so it prioritize the good then avg and then poor, if not assigned then it will select randomly
oe = OrdinalEncoder(categories = [['Poor','Average','Good'],['School','UG','PG']])
```

```python
# now to train:
oe.fit(x_train)
```

```python
# now its trained lets encode all:
x_train = oe.transform(x_train)
x_train
# we can compare it to previous x_train poor-0,school-0: 1st row
```

```python
# now till now it was ordinal encoder now lets see label encoder:
# label encoder is for yes no,off on,etc have same weight unlike ordinal
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train
```

```python
le.fit(y_train)
```

```python
le.classes_
```

```python
y_train = le.transform(y_train)
y_test = le.transform(y_test)
y_train
```

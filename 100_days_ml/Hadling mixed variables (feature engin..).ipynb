```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('titanic_changed.csv')
df.sample(5)
```

```python
# in this changed titanic set there is mixture of numeric and string in single columns:
# lets divide them for each categories:

# lets check what are unique sets in column: 'number'
df['number'].unique()
```

```python
# lets see how many each category have the data divided:

fig = df['number'].value_counts().plot.bar()
fig.set_title('Passengers travelling:')
# A for alone, 2,1,3,... are number they are travelling together
```

```python
# now lets extract numerical part:

df['number_numerical'] = pd.to_numeric(df['number'],errors = 'coerce', downcast = 'integer')

# now for categorical part:

df['number_categorical'] = np.where(df['number_numerical'].isnull(),df['number'],np.nan)

df.head()
```

```python
# now lets check for column: 'Cabin'
df['Cabin'].unique()
```

```python
 df['Ticket'].unique()
```

```python
# there is lot more unique value in these columns so lets again categorise them in num and categorical(strings):

df['cabin_num']= df['Cabin'].str.extract(r'(\d+)')  #sellect numerical part

df['cabin_cat'] = df['Cabin'].str[0]                #select the first letter

df.head()
```

```python
# now lets see the how much each of these new categories contain:
df['cabin_cat'].value_counts().plot(kind='bar')
```

```python
# now for ticket column:

# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])

df['ticket_num'] = pd.to_numeric(df['ticket_num'], errors = 'coerce', downcast = 'integer')


# extract the first part of ticket as category:
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])

df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan, df['ticket_cat'])

df.head(10)
```

```python
# after changing lets how many unique are:
df['ticket_cat'].unique()
```

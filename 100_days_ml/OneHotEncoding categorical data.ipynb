```python
# since there are lot of categories let make a new categores for brand having less than 100 cars:

counts = df.brand.value_counts()     # total no. of brand
threshold = 100
repl = counts[counts <= threshold].index    # making new for 100 and less into new category- repl
repl
```

```python
pd.get_dummies(df.brand.replace(repl,'uncommon'))
# the new column for repl is uncommon:
```

```python
# lets chack for less than 100 cars for uncommon:
pd.get_dummies(df['brand'].replace(repl,'uncommon')).sample(5)
```

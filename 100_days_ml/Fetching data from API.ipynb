```python
import pandas as pd
import requests
```

```python
# generated api key and api from tmdb:
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=175f80a73de21298b70e67b663040ace&language=en-US&page=1')
# sometime due to slow net its not working so we use timeout
response
```

```python
pd.DataFrame(response.json()['results']).head()
```

```python
# now extract the columns we want:
df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
df.head()
```

```python
# this is only page-1 and there are 500 such pages from that api, we have to add them all:
```

```python

df=pd.DataFrame()
for i in range(1,501):
  response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=175f80a73de21298b70e67b663040ace&language=en-US&page={}'.format(i))
  df2 = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
  df = pd.concat([df, df2], ignore_index=True)


df
```

```python
df.shape
```

```python
# save it in csv format:
df.to_csv('topRatedMovies.csv')
```

# categorical - categorical :- heatmap

```python
pd.crosstab(tt.Pclass,tt.Survived)
```

```python
sns.heatmap(pd.crosstab(tt.Pclass,tt.Survived))
```

```python
#must have three argument: values, index, columns: just like this:
flight.pivot_table('passengers','month','year') 
# or
flight.pivot_table(values = 'passengers',index = 'month',columns='year') 
```

```python
sns.heatmap(flight.pivot_table('passengers','month','year') )
# heat map is best when there are lot of numerical numerical comparison:
```

```python
sns.clustermap(flight.pivot_table('passengers','month','year') )
# this merges the ones with same behaviour
```

```python
tt.groupby('Pclass')['Survived'].mean()*100
```

```python
# its showing 62.9% in Pclass survived:
(tt.groupby('Pclass')['Survived'].mean()*100).plot(kind='bar')
```

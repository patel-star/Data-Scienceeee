```python
x=[0,1,2,3,4,5]
y=[0,1,4,9,16,25]

# This is long notations for formating the line drawn:
# plt.plot(x,y,label='x^2',color='red',linewidth='2',linestyle='--',marker='.',markersize=10,markeredgecolor='blue')

# This is short notations:
plt.plot(x,y,'b.--',label='x^2')

plt.title("square numbers",fontdict={'fontsize':18})
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.xticks([0,1,2,2.5,3,4,5,6]) #extra numberss on axis if u wanna
plt.yticks([0,1,4,9,16,25])

plt.legend()

plt.show()
```

```python
# Line-1
x=[0,1,2,3,4,5,6]
y=[0,3,6,9,12,15,18]

# Resize your graph
plt.figure(figsize=(10,5),dpi=300)

plt.plot(x,y,'r^--',label='3x')

#Line-2
x2=np.arange(0,5.5,0.5) #arange(start,stop,step) here excluding 5.5

# plt.plot(x2,x2**2,'b.-',label='x^2')
plt.plot(x2[:6],x2[:6]**2,'b',label='x^2') #here 6 is exclusive and only blue color is given so bydefault lline will be solid continious
plt.plot(x2[5:],x2[5:]**2,'b--')

plt.title("y=3x",fontdict={'fontsize':18})
plt.xlabel("x axis-x")
plt.ylabel("y axis-3x")

plt.legend()
plt.savefig('my-first-graph.png',dpi=300)

plt.show()
```

# ***Bar Chart:***

```python
labels=['A','B','C']
values=[11,88,44]

# a little bit different from ploting a line ie:
bars = plt.bar(labels,values)

# bars[0].set_hatch('/') #changee the first bar
# bars[1].set_hatch('*')
# bars[2].set_hatch('O')

# or we can change that with simple and sort:
patterns = ['/','*','O']
for bar in bars:
  bar.set_hatch(patterns.pop(0)) #here pop is used for pop / first then use it in bar from patterns


plt.show()

#set the size and  quality of graph image
plt.figure(figsize=(10,5),dpi=300)

plt.show()
```

# **Real world Examples:**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## *Line graph*

```python
gas = pd.read_csv('gas_prices.csv')
# print(gas.head())                             #to see the what are the contents

plt.figure(figsize=(8,5))                     #define the plot size
plt.title('Gas Prices over Years (in USD)',fontdict={'fontweight':'bold','fontsize':18})   #title of graph

plt.xlabel('Year')                            #label for x axis
plt.ylabel('US Dollars')                      #label for y axis

## this is for specifies countries:
plt.plot(gas.Year,gas.USA,'b.-',label='USA')  #plot the graph having marker and label defined
plt.plot(gas.Year,gas.UK,'g.-',label='UK')
plt.plot(gas.Year,gas['South Korea'],'r.-',label='South Korea')



plt.xticks(gas.Year[::3])                     #showing the years in x having 3 year difference

plt.legend()                                  #showing the legend, that is marker above

# save the graph:
plt.savefig('gas-prices.png',dpi=300)
plt.show()
```

```python
# this is for all countries in gas_prices.csv :

plt.figure(figsize=(10,8))

for column in gas:
  if column != 'Year':
    plt.plot(gas.Year,gas[column],marker='.',label=[column])

plt.legend()
plt.show()
```

```python
# plot for specific countries:
plt.figure(figsize=(10,8))

# these are the countries we want to look at, so we make list first:
countries_to_look_at = ['Germany','USA','Canada','Australia']

#then loop and plot all:
for country in countries_to_look_at:
  plt.plot(gas.Year,gas[country],marker='.',label=country)

plt.legend()
plt.show()
```
